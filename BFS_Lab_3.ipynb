```python
#Importing deque from collections:We import the deque data structure, which allows efficient appending and popping from both ends. This will be used for the BFS queue.
from collections import deque
#Function Definition:This defines the BFS function, which accepts a graph (in adjacency list format) and a start_node from where BFS traversal will begin.
def bfs(graph, start_node):
    # Initialize a queue to manage the nodes to be explored
    queue = deque([start_node])

    # Keep track of the visited nodes to avoid loops
    #Visited Set:
    visited = set([start_node])

    # List to store the order of nodes visited:This list will store the order in which the nodes are visited during BFS.
    bfs_order = []

    # Loop until there are no more nodes to explore
    while queue:
        # Dequeue the first node from the queue
        #The current_node is dequeued (removed from the front of the queue) and stored for processing.
        current_node = queue.popleft()

        # Append the current node to the bfs_order list:The current_node is added to the list bfs_order to track the traversal order.
        bfs_order.append(current_node)
        #Exploring Neighbors:
        # This loop goes through all the neighbors of the current_node as defined in the adjacency list (graph).
        for neighbor in graph[current_node]:
            # If the neighbor hasn't been visited yet
            if neighbor not in visited:
                # Add the neighbor to the visited set
                visited.add(neighbor)
                # Enqueue the neighbor for future exploration:neighbor is added to the queue for future exploration.
                queue.append(neighbor)
    #Once the loop completes, the BFS traversal order is returned.
    return bfs_order

# Define the COMSATS graph
comsats_graph = {
    'CUI Islamabad': ['CUI Lahore', 'CUI Abbottabad', 'CUI Vehari'],
    'CUI Lahore': ['CUI Islamabad', 'CUI Abbottabad', 'CUI Wah'],
    'CUI Abbottabad': ['CUI Islamabad', 'CUI Lahore', 'CUI Wah'],
    'CUI Vehari': ['CUI Islamabad', 'CUI Wah'],
    'CUI Wah': ['CUI Lahore', 'CUI Abbottabad', 'CUI Vehari']
}

# Perform BFS starting from 'CUI Islamabad'#Calling BFS:
bfs_result = bfs(comsats_graph, 'CUI Islamabad')

# Print the result
print("Breadth-First Search Order:", bfs_result)
```
